# 1. Problem Description

This competition is (Jang, Uzsoy, and Culliton 2020) and it is about
using a Generative Adversarial Network neural network architecture to
add an artistic style to images based on an artist (Monet) images.

The workflow for this research is summarized in
<a href="#fig-proj-work" class="quarto-xref">Figure 1</a>. The process
begins with exploratory data analysis and preprocessing, followed by
model training. Hyperparameter tuning is iteratively applied to refine
model performance, culminating in the development of final models for
Kaggle submission.
